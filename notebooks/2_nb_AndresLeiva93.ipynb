```python
import pandas as pd
import numpy as np
```

```python
dolar_2019 = pd.read_csv('../data/raw/dolar_2019_sii.csv',sep=';', nrows=30)
dolar_2019.head()
```

Adicional al nombre del df, el argumento necesario para leer el archivo es el separador, en este caso el ';', utilizamos el parametro 'nrows' para leer solo las filas que tienen datos y no el total.

```python
dolar_2020_txt = pd.read_csv('../data/raw/dolar_2020_sii.txt',sep='|', nrows=30,header=None)
dolar_2020_txt.head()
```

Con respecto al df anterior, este se diferencia en que el tipo de separador ya no es el ';', sino que es el '|' y en que no contiene los encabezados. Son iguales en que la columnas están igualmente estructuradas, ambas representan los meses. Para la lectura correcta del df es necesario agregar el separador y definir con header=None para que no utilice la primera fila como columna.

```python
import re
dolar_2020_regex = pd.read_csv('../data/raw/dolar_2020_sii_AA.dat',sep="[a-zA-Zá-ý]{3}", nrows=30,header=None)
dolar_2020_regex.head()
```

Este df tiene la misma información que el anterior con el único detalle que está separado con caracteres especiales.

```python
#homogeneizmos los encabezados para los tres dataframe
dolar_2020_txt.columns = dolar_2019.columns
dolar_2020_regex.columns = dolar_2019.columns
```

```python
#Crea columna year con type int64
dolar_2019['year'] = np.int64(2019)
dolar_2020_txt['year'] = np.int64(2020)
dolar_2020_regex['year'] = np.int64(2020)
```

```python
#Cambamos . por ,
for c in dolar_2019.iloc[:,1:13].columns:
    dolar_2019[c] = dolar_2019[c].str.replace(',', '.').astype(float)
for c in dolar_2020_txt.iloc[:,1:13].columns:
    dolar_2020_txt[c] = dolar_2020_txt[c].str.replace(',', '.').astype(float)
for c in dolar_2019.iloc[:,1:13].columns:
    dolar_2020_regex[c] = dolar_2020_regex[c].str.replace(',', '.').astype(float)
```

```python
dolar_2019.to_csv("../data/interim/d2019.csv", encoding="latin1",index=False)
dolar_2020_txt.to_csv("../data/interim/d2020.txt", encoding="latin1",index=False)
dolar_2020_regex.to_csv("../data/interim/dolar_2020_excel.csv", encoding="latin1",index=False,sep=';')
```

```python
df_concat_1 = pd.concat([dolar_2019,dolar_2020_txt],axis=1)
```
